```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(values ='salary_year_avg', index ='job_country', columns = 'job_title_short', aggfunc ='median')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
df_job_country_salary = df.pivot_table (
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($US)')
plt.xlabel('')
plt.title('Median Salary By Country And Job Title')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
pivot_table = df.pivot_table(index=['job_title_short', 'job_country'], aggfunc='size')
pivot_table.head(10)
```

```python
pivot_table_2 = df.pivot_table(
    values='salary_year_avg',
    index='company_name',
    aggfunc=['min','max','mean']
    )
pivot_table_2.head(5)
```

```python
pivot_table_3 = df[df['job_title_short'] == 'Data Scientist'].pivot_table(
    values='salary_year_avg',
    index=['company_name','job_country'],
    aggfunc='median'
)

pivot_table_3 = pivot_table_3[pivot_table_3 > 200000]

pivot_table_3.head(10)

```
