```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US_jobs =df[df['job_country']=='United States']
```

```python
df_US_jobs['job_posted_month']= df['job_posted_date'].dt.strftime('%B')
```

```python
df_monthly_postings = df_US_jobs.groupby('job_posted_month').size().reset_index(name='num_postings')
```

```python
merged_df =pd.merge(df_US_jobs,df_monthly_postings, on='job_posted_month',how='right')
```

```python
df_not_remote_job = df[df['job_work_from_home']==False]
df_not_remote_salary= df_not_remote_job.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary.rename(columns={'salary_year_avg':'salary_not_remote_avg'},inplace=True)
merged_df_2=pd.merge(df_not_remote_job,df_not_remote_salary,on='job_title_short',how='right')
```

```python
merged_df_2.head()
```

```python
df_company_salary=df.groupby('company_name')['salary_year_avg'].mean().reset_index(name='avg_salary')
df_company_remote=df[df['job_work_from_home']==True]
df_company_remote=df_company_remote.groupby('company_name').size().reset_index(name='remote_count')
merged_df_3=pd.merge(df_company_salary,df_company_remote,on='company_name',how='inner')
```
